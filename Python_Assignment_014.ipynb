## 1. What does RGBA stand for?

RGBA stands for "Red, Green, Blue, Alpha". It is a color model used in computer graphics and design to represent colors as a combination of red, green, and blue light, along with an alpha channel for transparency.

The RGBA model is similar to the RGB model, but includes an additional alpha channel that specifies the opacity of the color. The alpha channel ranges from 0 (fully transparent) to 1 (fully opaque).

## 2. From the Pillow module, how do you get the RGBA value of any images?

```python
from PIL import Image   #Python Imaging Library - Pillow module

# Open an image file
image = Image.open("RGBA.jpg")

# Get the RGBA value of a pixel at position (x, y)
x, y = 10, 20
rgba = image.getpixel((x, y))

# Print the RGBA value
print(rgba)
```

## 3. What is a box tuple, and how does it work?

In Pillow, a box tuple is a tuple of four integers that represents a rectangular region in an image. The four integers represent the left, upper, right, and lower coordinates of the rectangular region, respectively.

The box tuple is often used as an argument in various methods of the Image class, such as the crop() method or the resize() method, to specify a rectangular region of an image to be cropped or resized.

```python
from PIL import Image

# Open an image file
image = Image.open("RGBA.jpg")

# Define a box tuple representing the rectangular region to be cropped
box = (100, 100, 300, 300)  # (left, upper, right, lower)

# Crop the image to the specified rectangular region
cropped_image = image.crop(box)

# Save the cropped image to a file
cropped_image.save("cropped_image.png")
```

## 4. Use your image and load in notebook then, How can you find out the width and height of an Image object?

```python
from PIL import Image

# Open an image file
image = Image.open("RGBA.jpg")

# Get the width and height of the image
width, height = image.size

# Print the width and height
print(f"Width: {width}, Height: {height}")
```

## 5. What method would you call to get Image object for a 100×100 image, excluding the lower-left quarter of it?

```python
from PIL import Image

# Open the original image
image = Image.open("RGBA.jpg")

# Define the box tuple for the upper-right quadrant
box = (50, 0, 100, 50)  # (left, upper, right, lower)

# Crop the image to the upper-right quadrant
cropped_image = image.crop(box)

cropped_image.save("cropped_image.png")

# Resize the cropped image to 100x100
resized_image = cropped_image.resize((100, 100))

# Save the resized image
resized_image.save("resized_image.png")
```

## 7. What module contains Pillow’s shape-drawing code?

```python
from PIL import Image, ImageDraw

# Open an image file
image = Image.open('nature.jpg')

# Create an ImageDraw object
draw = ImageDraw.Draw(image)

# Draw a red line from (0, 0) to (100, 100)
draw.line((0, 0, 100, 100), fill='red')

# Save the modified image
image.save('modified_image_nature.jpg')
```

## 8. Image objects do not have drawing methods. What kind of object does? How do you get this kind of object?

The Image objects in Pillow do not have direct drawing methods, but you can use the ImageDraw module to create an object that has drawing methods. The ImageDraw module provides the Draw class which allows you to draw shapes, text, and other graphical elements onto an image.

```python
from PIL import Image, ImageDraw

# Open an image file
image = Image.open('nature.jpg')

# Create an ImageDraw object
draw = ImageDraw.Draw(image)

# Draw a circle on the image
draw.ellipse((50, 50, 150, 150), outline='red')

# Save the modified image
image.save('modified_example.jpg')
```
