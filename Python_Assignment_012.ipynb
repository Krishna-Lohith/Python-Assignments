### 1. In what modes should the PdfFileReader() and PdfFileWriter() File objects will be opened?

**PdfReader() File object mode: 'rb'**
The PdfFileReader() class is used for reading PDF files. When creating an instance of PdfFileReader(), you should open the PDF file in binary read mode ('rb'), which allows reading the file as binary data.

```python
from PyPDF2 import PdfReader

# Open PDF file in binary read mode
with open('statistics.pdf', 'rb') as file:
    pdf_reader = PdfReader(file)
    print(pdf_reader)
    # Perform operations with pdf_reader object
```

**PdfWriter() File object mode: 'wb'**
The PdfFileWriter() class is used for writing PDF files. When creating an instance of PdfFileWriter(), you should open the PDF file in binary write mode ('wb'), which allows writing binary data to the file.

```python
from PyPDF2 import PdfWriter 

#Truncates whole pdf data and start writing new data.
with open('statistics.pdf', 'wb') as file:
    pdf_writer = PdfWriter()
    pdf_writer.write(file)
```

### 2. From a PdfFileReader object, how do you get a Page object for page 5?

```python
from PyPDF2 import PdfReader

# Open PDF file in binary read mode
with open('statistics.pdf', 'rb') as file:
    pdf_reader = PdfReader(file)
    # Get Page object for page 5 (page index is 0-based)
    page_5 = pdf_reader.pages[5]  
    # Note: Page index starts from 0, so page 5 corresponds to index 4

    # Perform operations with the Page object
    print(f"Page 5 contents: {page_5.extract_text()}")
```

### 4. If a PdfFileReader object’s PDF is encrypted with the password swordfish, what must you do before you can obtain Page objects from it?

```python
from PyPDF2 import PdfReader

with open('statistics Encrypted(pass-statistics).pdf', 'rb') as file:
    pdf_reader = PdfReader(file)

    if pdf_reader.is_encrypted :
        pdf_reader.decrypt('statistics')
        
    num_pages = len(pdf_reader.pages)

    for page_num in range(num_pages):
        page = pdf_reader.pages[page_num]
        print(page.extract_text())
```

### 5. What methods do you use to rotate a page?

```python
from PyPDF2 import PdfReader, PdfWriter

# Open the input PDF file
with open('statistics.pdf', 'rb') as input_file:
    # Create a PdfReader object
    pdf_reader = PdfReader(input_file)

    # Get the first page of the PDF
    page = pdf_reader.pages[0]

    # Rotate the page clockwise by 90 degrees
    page.rotate(90)

    # Create a PdfWriter object
    pdf_writer = PdfWriter()

    # Add the rotated page to the PdfWriter
    pdf_writer.add_page(page)

    # Create the output PDF file
    with open('rotated statistics.pdf', 'wb') as output_file:
        # Write the PdfWriter object to the output file
        pdf_writer.write(output_file)
```

### 6. What is the difference between a Run object and a Paragraph object?

In the context of text processing using Python libraries such as python-docx or python-docx2pdf, a "Run" object and a "Paragraph" object refer to different elements in a document.

A "Run" object represents a contiguous sequence of text with the same formatting properties within a paragraph. It can include plain text, as well as text with specific formatting such as font size, font style (bold, italic, underline), color, and more. Runs are used to apply formatting to specific portions of text within a paragraph.

A "Paragraph" object, on the other hand, represents a block of text in a document. It is typically a complete line of text that ends with a paragraph mark. A paragraph can consist of one or more "Run" objects, and it is used to define the overall formatting properties of the text block, such as alignment, indentation, spacing, and more.

In summary, a "Run" object represents a portion of text with specific formatting properties within a paragraph, while a "Paragraph" object represents a block of text with overall formatting properties for that text block. "Runs" are used for applying formatting to specific portions of text, while "Paragraphs" are used for defining the formatting properties of a text block as a whole.

### 9. What is the difference between False, True, and None for the bold variable?

In the context of document processing libraries, such as python-docx or python-docx2pdf, which are used for working with Microsoft Word documents in Python, the bold variable is a boolean attribute that represents whether a text run is formatted as bold or not.

The values False, True, and None for the bold variable represent different states of text formatting as follows:

False: When bold is set to False, it means that the text run is not formatted as bold. The text will be displayed in the regular (non-bold) font style.

True: When bold is set to True, it means that the text run is formatted as bold. The text will be displayed in a bold font style.

None: When bold is set to None, it means that the text run does not have an explicitly defined bold formatting. The formatting will be inherited from the style applied to the paragraph or the document level. If no style is applied at the paragraph or document level, the text will be displayed in the regular (non-bold) font style.

In summary, False indicates that the text is not bold, True indicates that the text is bold, and None indicates that the text inherits its bold formatting from the paragraph or document level style.
